```python
import xml.etree.ElementTree as ET
import re
```

```python
# READING TEXT (PARAGRAPHS ONLY) FROM XML WITH SUPER/SUB TAGS

tree = ET.parse('Klemm et al.tei.xmlScitex.xml')
root = tree.getroot()
text = ''

for elem in root:
    if (elem.tag[29:] == 'text'):
        for sub_elem1 in elem:
            if (sub_elem1.tag[29:] == 'body'):
                for sub_elem2 in sub_elem1:
                    if sub_elem2.tag[29:] == 'div':
                        for sub_elem3 in sub_elem2:
                            if sub_elem3.tag[29:] == 'p':
                                xml_str = ET.tostring(sub_elem3, encoding='unicode')
                                text = text + xml_str
                                
pattern_starting_p_tag = r'<ns0:p[^>]+>'
pattern_ending_p_tag = '</ns0:p>'
pattern_starting_ref_tag = r'<ns0:ref[^>]+>'
pattern_ending_ref_tag = '</ns0:ref>'

text = re.sub(pattern_starting_p_tag, '', text)
text = re.sub(pattern_ending_p_tag, '', text)
text = re.sub(pattern_starting_ref_tag, '', text)
text = re.sub(pattern_ending_ref_tag, '', text)

text
```

```python
del text
```

```python
# READING TEXT (PARAGRAPHS ONLY) FROM XML WITHOUT SUPER/SUB TAGS 

tree = ET.parse('Klemm et al.tei.xmlScitex.xml')
root = tree.getroot()
text = ''

for elem in root:
    if (elem.tag[29:] == 'text'):
        for sub_elem1 in elem:
            if (sub_elem1.tag[29:] == 'body'):
                for sub_elem2 in sub_elem1:
                    if sub_elem2.tag[29:] == 'div':
                        for sub_elem3 in sub_elem2:
                            if sub_elem3.tag[29:] == 'p':
                                for sub_elem4 in sub_elem3.itertext():
                                    text += sub_elem4

text
```
